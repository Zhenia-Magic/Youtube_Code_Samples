```python
import pandas as pd
df = pd.read_csv("train-2.csv")
df
```

```python
df.pivot_table(index=['Survived'])
```

```python
df.pivot_table(index=['Survived', 'Sex'], values = 'Fare')
```

```python
df.pivot_table(index='Survived', columns = 'Sex', values = 'Fare', aggfunc = ['max', 'mean'])
```

```python
def range_function(series):
    return max(series) - min(series)

df.pivot_table(index='Survived', columns = 'Sex', values = 'Fare', aggfunc = range_function)
```

```python
df.pivot_table(index=['Pclass','Sex'],  aggfunc = {'Fare': ['min','max'], 'Age' : 'mean'})
```

```python
df.pivot_table(index='Survived', columns = 'Sex', values = 'Fare', aggfunc='count', margins = True)
```
